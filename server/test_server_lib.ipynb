### Temporal requests

```python
print('Testing temporal requests...')

all_regions = Regions.get_regions('ES')
print(all_regions)
datasources = COnVIDa.get_data_items_names(DataType.TEMPORAL, language='EN')
all_data_items = []
for data_items in datasources.values():
    all_data_items += data_items
    
start_date = pd.to_datetime('2020-11-01', format='%Y-%m-%d')
print('START DATE: ', start_date)
end_date = pd.to_datetime('2020-12-30', format='%Y-%m-%d')
print('END DATE: ', end_date)
```

```python
data = convida_server.get_data_items(data_items=[all_data_items[0]],
                              regions=all_regions,
                              start_date=start_date,
                              end_date=end_date,
                              language='EN')
display(data.head(2))
display(data.tail(10))
```
